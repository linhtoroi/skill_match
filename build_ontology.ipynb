```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

temp = []
def flatten(d_in, base=()):
    for k in d_in:
        if type(d_in[k]) == dict:
            flatten(d_in[k], base+(k,))
            # print(k, d_in[k].keys())
            if len(d_in[k].keys())>0:
                for i in d_in[k].keys():
                    temp.append({'from' :k,'to': i})
                    # if 'computer programming language' in i:
                    #     print(i)
                    #     print(d_in[k][i])
                    #     print(k)
        else:
            print(base + (k, d_in[k]))

flatten(g_onto)

df = pd.DataFrame(temp)
df=df.drop_duplicates()
G = nx.from_pandas_edgelist(df, 'from', 'to')

def draw_graph(G):
    plt.figure(figsize=(12,12))
    pos = nx.spring_layout(G)
    nx.draw(G, with_labels=True, node_color='skyblue', edge_cmap=plt.cm.Blues, pos = pos, node_size = 4500)
    plt.show()

draw_graph(G)
```
