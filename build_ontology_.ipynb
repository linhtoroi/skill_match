```python
import os
import pickle
path = './ontology/ontology/'
skill_ontology = {}
for file in os.listdir(path):
    if file.endswith(".p"):
        with open(os.path.join(path, file), 'rb') as f:
            try:
                while True:
                    skill_ontology[file.replace('_ontology.p','')]=pickle.load(f)
            except EOFError:
                pass
```
